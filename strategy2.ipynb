```python
import pandas as pd

df = pd.read_csv("BTC-USDT_daily.csv")

# Find the maximum value from the first day's high
max_value = df['high'].iloc[0]

# Find the minimum value from the second day's low
min_value = df['low'].iloc[0]

df['max_value'] = None
df['min_value'] = None
df.loc[0, 'max_value'] = df['high'].iloc[0]
df.loc[0, 'min_value'] = df['low'].iloc[0]
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df['signal'] = None
df.loc[0, 'signal'] = 0
# Loop through the DataFrame from the second day onwards
for i in range(1, len(df)):
    # Check if the high value is greater than the current interval max
    if df['close'].iloc[i] > df.loc[i-1, 'max_value']:
        # Update the maximum value of the interval range
        df.loc[i, 'max_value'] = df['high'].iloc[i]
        df.loc[i, 'min_value'] = df['low'].iloc[i]
        df.loc[i, 'signal'] = 0
    # Check if the low value is less than the current interval min
    elif df['close'].iloc[i] < df.loc[i-1, 'min_value']:
        # Update the minimum value of the interval range
        df.loc[i, 'max_value'] = df['high'].iloc[i]
        df.loc[i, 'min_value'] = df['low'].iloc[i]
        df.loc[i, 'signal'] = 0

    else:
        df.loc[i, 'max_value'] = df['max_value'].iloc[i-1]
        df.loc[i, 'min_value'] = df['min_value'].iloc[i-1]
        df.loc[i, 'signal'] =  df.loc[i-1, 'signal'] + 1
```

```python
df
```

```python
df['signal_diff'] = df['signal'].diff()
df['buyORsell'] = 0  
for index, row in df.iterrows():
    if row['signal_diff'] <= -3 and row['close'] > df['max_value'].iloc[index-1]:
        df.loc[index, 'buyORsell'] = 1 
    elif row['signal_diff'] <= -3 and row['close'] < df['min_value'].iloc[index-1]:
        df.loc[index, 'buyORsell'] = -1 
    else: df.loc[index, 'buyORsell'] = 0
        
```
