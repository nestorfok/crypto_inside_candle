```python
import pandas as pd
```

```python
benchmark_performance = pd.read_csv('benchmark_performance.csv')
```

```python

# btc = benchmark_performance[benchmark_performance["sym.2"] == "BTC"]
benchmark_performance.drop(['sym', 'returns', 'Unnamed: 2', 'sym.1', 'year', 'returns.1', 'Unnamed: 6'], axis=1, inplace=True)
benchmark_performance = benchmark_performance.reset_index(drop=True)
benchmark_performance
```

```python
benchmark_performance['month'] = benchmark_performance['month'].astype(str)

for i in range(len(benchmark_performance)):
    if benchmark_performance.iloc[i]['month'].endswith(".1"):
        benchmark_performance.at[i, 'month'] = benchmark_performance.iloc[i]['month'] + "0"

benchmark_performance['month'] = pd.to_datetime(benchmark_performance['month'], format="%Y.%m")
benchmark_performance
```

```python
benchmark_performance.to_csv("benchmark.csv")
```
